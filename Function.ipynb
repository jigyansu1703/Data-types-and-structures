```python
#1. Write a python function that takes a list of numbers as input and returns the sum of all positive numbers in the list.
def sum_of_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum
```

```python
my_list = [1,2,3,4,5,6]
result = sum_of_even_numbers (my_list)
print("sum_of_even_numbers",result)
```

```python
#4. Write a python function that checks if a given number is prime or not from 1 to 200
def is_prime(n):
  if n<2 or n>200:
    return False
  for i in range(2,int(n**0.5)+1):
    if n%i ==0:
      return False
  return True
```

```python
# 5. Create an iterator class in python that generates the Fabonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms  # Total number of terms to generate
        self.count = 0              # Counter to keep track of number of terms generated
        self.a = 0                  # First Fibonacci number
        self.b = 1                  # Second Fibonacci number

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # Stop iteration after reaching max terms
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.count += 1
            return next_value
```

```python
fib = FibonacciIterator(15)  # Generate first 10 Fibonacci numbers

for num in fib:
    print(num, end=' ')
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string
def read_file_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # remove newline character
    except FileNotFoundError:
        print("File not found.")
# Replace 'example.txt' with your actual file path
for line in read_file_lines('example.txt'):
    print(line)
```

```python
# 9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
words = ("a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z")
vowels = ('a','e','i','o','u')
remove_vowels = list(filter(lambda x: x not in vowels,words))
print(remove_vowels)
```
